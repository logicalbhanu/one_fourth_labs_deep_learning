*An exper system is also known as rule based system*

**Binary Classification :** any problem of answer type in "yes or NO" format is called a binary classification eg- a given image contains text or not? , now this question can be answered in Yes or No format so it is a binary classification problem.

**Human Decision Making** : since we humans whenever dealt with any Binary classification problem we try to solve those problems by bifurcating that problem into a set which contains its causes which led to decide whether the problem should be answered in Yes or No and a set which contains its result i.e Yes or No.
eg: Say A doctor decides whether a person has dengue or not by considering certain symptoms which are: *Skin Rash, Fever, Headache, Cold cough, Vomiting* which finally led to decide him whether patient has dengue or not.
now here the symptoms can be reffered as causes set and the result can be interpret as result set.we represent Yes with 1 and no with 0 to convert the input into machine readable input.

**Features and Rules :** *Features* are all these inputs that we are taking in the above example.
*Rules* are the logics that the doctor used in his mind to decide that whether the patient has dengue or not.
note that the rule can be expressed as a set of if and else conditions which on feeding a input decides whether a person has dengue or not.

**Expert Systems** : an expert system is basically that rule in form of if and else conditions which on feeding the input would decide whether the patient has dengue or not, that expert system can be outsourced to a machine to let that machine decide whether a patient has dengue or not on feeding the corresponding inputs.
