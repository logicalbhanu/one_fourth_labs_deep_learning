# The python classes


```
class MobilePhone:
  """ This is what we called the docstring and it is  only for givig the textual description to the class defined and it can be
  accessed through using the inbuilt docstring variable of the class"""
  def __init__(self,name,is_android=False,screen_size=4.3):
    self.name = name
    self.is_android=is_android
    self.rating = -1
    self.screen_size= screen_size
  def has_ratings(self):
    return self.rating > -1
```

```
new_phone = MobilePhone('Aex_rom')
```

```
type(new_phone)
```

```
new_phone.has_ratings()
```

```
new_phone.rating=2.2
new_phone.has_ratings()
```

# The Inheritance concept

```
class iphone(MobilePhone):
  def __init__(self,name):
    MobilePhone.__init__(self,name,False,4)
    print("this is me and this is always me forever")
```

```
new_iphone = iphone('iphone 10')
```

```
new_iphone.is_android
```

```
iphone.__init__(new_phone,"the foolish")
```

```
MobilePhone.has_ratings() #here we have to pass a class object as parameter to remove the error
```

```
MobilePhone.has_ratings(new_iphone) # see programiz 'class' tutorial to have a better grasp
# in a regular object created by 'MobilePhone()' as constructor say "ob=MobilePhone()", calling a method 
# say 'ob.has_ratings()' is equal to 'ob.has_ratings(ob)' to be compatible with 'self' argument in method prototype whereas
# it is not similar with object which get created having name same as the class name when the class get created i.e 'MobilePhone'
# hence we have to pass the class object while calling any method through it.
```

# The Basic Plotting

```
import matplotlib.pyplot as plt
import numpy as np
```

```
x = np.random.random((10,1))
print(x)
```

```
plt.plot(x)
plt.show()
```

```
# here in the above graph list indexes has been treated as x-coordinates
# and the values of the element at that indexes has been treated as y-coordinates
```

```
plt.plot(x,'*')
plt.show() # here the values has been shown as '*' i.e y coordinates
```

```
plt.plot(x,'*-')
plt.show() # this is to use both the '*' and '-'(line) for plotting
# this also suggest that we have just pass the symbols as string to use 
# them for plotting
```

```
plt.plot(x,'+')
plt.show()
```

```
x = np.linspace(0,10,100) # here 100 is the no of steps to be inserted 
# between 0 and 10
# to change the spacing either we have to make partitions small i.e change
# the value(decrease) of 100 or increase the limit of 10
y = np.power(x,0.5) # np.power will calculate the 0.5 power(square root) of all the 
# of all the elements of np-array x
```

```
plt.plot(x,y) # passing the two arguments make x as x-coordinates and
              # y as y-coordinates
plt.show()
```

```
plt.plot(x,y,'*')
plt.show()
```

```
plt.plot(x,y,'*-')
plt.show()
```
