# Euclidean space :
 In geometry, a two or three-dimensional space in which the axioms and postulates of  Euclidean geometry apply is an Euclidean space. A space in any finite number of dimensions, in which points are designated by coordinates (one for each dimension) and the distance between two points is given be a distance formula.

 basically any space having any number of dimensions  where points are represented by its co-ordinates and distance between the points is represented by a distance formula is called an Euclidean Space.

l2-norm is also known as eculidean norm is basically the way we calclulate the distance formula for distances between the two points in an euclidean plane.

***Orthogonal Vectors:*** Vectors perpendicular to each other are called orthogonal vectors and to check them the necessary and sufficient condition is that there dot product would be zero.

***note: two vectors will have the same direction if one can be expressed as a multiple of the other. Note that [0.9, 1.5, 2.1] = 0.6 * [1.5, 2.5, 3.5] ,therefore [0.9, 1.5, 2.1] and [1.5, 2.5, 3.5] have same direction or the angle between them is zero.***

***note:Remember that the projection of a on b will be in the direction of b. In other words, it will have the same direction as b but its magnitude will be different. In other words, it will be a multiple of b.***

# Why vectors: 
    Now the question is, what is the use of vectors in machine learning and how does it creating a difference in machine learning at all?
    Since the logic behind this is the possibility of representing anything using vectors, for example a product can also be represented as a vector, to represent a product say a mobile phone as a vector what we have to do is to represent all its attributes such as it memory, battery , ratings and other things as various coordinates of vector, even thing or the attribute whose value is textual in nature can also be represented as vector coordinates by converting these values into numbers for example 'is android or not' feature can be recorded as binary value in the coordinates of vector(1 for android and 0 for not being android phone).
    So this shows that everything can be represented as vectors with a bit creativity applied.

    Now the next question is, so what if anything can be represented as vector how is it usefull at all?
    This can be answered as an example, say we calculate the angle between the two vectors representing the two products. now if the angle between them is less then this means that the product are similar in nature as there are pointing towards the same coordinates and if not the same coordinates then in the coordinates which are closer to each other and since each coordinates represent the product feature so it means these products have similar or closed features to each other.


    so this is how the vectors role comes into play in machine learning.
