*An exper system is also known as rule based system*

**Binary Classification :** any problem of answer type in "yes or NO" format is called a binary classification eg- a given image contains text or not? , now this question can be answered in Yes or No format so it is a binary classification problem.

**Human Decision Making** : since we humans whenever dealt with any Binary classification problem we try to solve those problems by bifurcating that problem into a set which contains its causes which led to decide whether the problem should be answered in Yes or No and a set which contains its result i.e Yes or No.
eg: Say A doctor decides whether a person has dengue or not by considering certain symptoms which are: *Skin Rash, Fever, Headache, Cold cough, Vomiting* which finally led to decide him whether patient has dengue or not.
now here the symptoms can be reffered as causes set and the result can be interpret as result set.we represent Yes with 1 and no with 0 to convert the input into machine readable input.

**Features and Rules :** *Features* are all these inputs that we are taking in the above example.
*Rules* are the logics that the doctor used in his mind to decide that whether the patient has dengue or not.
note that the rule can be expressed as a set of if and else conditions which on feeding a input decides whether a person has dengue or not.

**Expert Systems** : an expert system is basically that rule in form of if and else conditions which on feeding the input would decide whether the patient has dengue or not, that expert system can be outsourced to a machine to let that machine decide whether a patient has dengue or not on feeding the corresponding inputs.

# Jar 4 ~ Loss Function

Loss function is basically a function which tell how better a model is as compared to other model by operating on each point of given data and comparing the average difference between all these data points outputs and outputs obtained by the model itself. there are various type of loss function and in this course we are going to study three types of loss function: ***Square Error Loss, Cross Entropy Loss,and KL Divergence***

# Jar 5 ~ Learning Algorithm

Since to know the value of the parameters we actually turn the give problem ( or our task problem ) into the optimization problem by using the loss function as we convert our problem into a loss function and try to optimize the loss function.

Now to optimize the loss function we need to get that parameters which minimise our loss function or we have to learn these parameters.

To learn these parameters we need some algorithm rather than trying to guess them(called as Brute force way) such as gradient decent, and many more. These Algorithms comes under the category of Jar 5 ~ Learning Algorithm.

# Jar 6 ~ Evaluation

In evaluation what we have to do is to calculate how accurate our model is in such terms which can be easily interpreted by a layman and one of the way is called as evaluation

***Acurracy :*** It is calculated by generating the output for each instance of a given test data whose actual output is already known. 
After this accuracy is evaluated as follows : 
 
 accuracy = (no of isntances for which the output and true output are same) / (total no of outputs) * 100


the above one is just one of the way to calculate the accuracy of the model there are others such as : ***Top 'k' chances*** ( that is in given k chances, model should atleast results true output once ) eg: google search engine

there are other ways too, such as ***'precision of a model'  , 'recall of a model'***

Precision of a model : Now suppose we have a model to drive the outmated car. it is expected by the model to apply the brake whenever a obstacle is in front of the car, say the total number of times it applies the brakes be 'n', and total no times when a obstacle is in front of the car be 'm', now let the no of times  car applies the brake when an obastacle is actually in front of the car be 'p'. 

***Then the precision would be = p/n***

i.e  

 ***(no of times car applies the brakes when an obstacle is in front of it) / (total no of times it applies the brakes)***


**Recall of the Model : for the above case the recall of the model  would be = p/m**

i.e   

  **recall = (total no of the times of car applies the brakes when a obstacle is in front of it) / (total number of times an obstacle is in front of it)**

**note that 'loss function' works over on training data and 'evaluation' works over the test data.**
