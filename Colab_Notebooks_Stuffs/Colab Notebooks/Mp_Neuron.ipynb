**In Mp neuron model we have :**
 
 1)Input = {0,1} i.e the boolean set of numbers.

moreover we can also consider our f(g(x)) as the function which fires **one when g(x)-b >= 0** and it fires ***zero when g(x)-b < 0*** (here we only transfer the parameter 'b' to the other side of the inequality to form the new equation which presents parameter 'b' in much more usual form of mathematics and hence easily interpretable )

# Mp Neuron Loss

To calculate the loss function we have to know our estimated function through which we can calculate the loss between our estimated value and our deduced value.

note that loss can be expressed in many ways which represent somehow the loss or difference between the calculated result and actual result, here in mp neuron we choose our loss function as the square of the difference between the actual and deduced value.

why square of the difference: because the loss can be negative and positive which can led to cancel the loss in total which is not desirable.

Then why not |difference between the actual value and calculated value|
i.e absolute value of the difference:  Because the |x| function is non differentiable and hence cannot be used to find the parameter for which we would have the OPTIMUM LOSS(MINIMUM LOSS).

The geometric interpretation basically elaborates the weakness of MP neuron model in a manner stated below:

suppose we have a dataset which have two attributes say x and y then a possible function to model this problem would be ***x + y >= c
where 'c' is the parameter whoose value we have to determine or say we have already decide a value and called it as 'c' itself then if we plot this line on the graph then it create two regions on the graph by dividing the 2D plane into two parts,***

clearly all the points which would lie on the line would satisfy our model or we can say that these points are basically training examples and they are satisfying our model , in other words our model validates or perform well for atleast these sets of training examples.

moreover all the training examples which are ***satisfying the x+y>c part of the model*** function would lie in one of the regions created by the line on the graph plane, in other words all the points which lies in ***region represented by x+y>c*** equation are the training examples which suits or satisfy our model.

in short we have huge number of points or training examples(lying on the line and the one of the region) which satisfy our model or which can be predicted by our model accurately.

###***so here we can say that our model suits well for many number of training examples, but this is not the case with Mp neuron model, 
### let see why?***

 Now since earlier we had taken an example in which the features or attributes of the model can taken any real value in the given plane or in the 2D plane i.e we can have various values for x and y both  which results to huge number of training examples satisfying our modeled equation or function which in turn prove the usefullness of our model, but this is not the case with Mp neuron's features or attributes or input variable as they can basically take only two possible discrete values and that is {0,1}
 so for a model having two features only the possible set of point that total features can represents are ***{(0,1), (1,1), (1,0), (0,0)}
  so if we draw a line or represents the Mp neurons equation on a 2D graph(which is in this case = x+y>=b )***  could only be allowed to paas through these points as these are the only possible sets to output that can be generated from the features set, and since a line only pass through two distincts points atmost (if all these four points are not collinear), so it damn sure that our line whatever we choose based on the parameter 'b' is only going to satisfy the two training examples and ***leaving the rest of the training examples which is contributing the 50% of the total training examples in this case(i.e having two attributes only).***


 That is our model using Mp neuron concept can only have 50% of accuracy atmost for a problem having two feature variables only.
 hence it raise question regarding their efficiency which gives rise to other machine learning models.
